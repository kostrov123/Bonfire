```python
### YOUR CODE HERE ###
# Подгружаем необходмые бибдиотеки
from textwrap import wrap
import re
import datetime
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import missingno as msno
from pandas.tseries.offsets import MonthEnd, MonthBegin
from collections import Counter
```

```python
df_courses = pd.read_csv('Data/courses.csv',index_col = 0)# Читаем файл courses.csv
df_courses.rename(columns={'id':'course_id'}, inplace=True)

df_course_contents = pd.read_csv('Data/course_contents.csv')# Читаем файл course_contents.csv

df_progresses = pd.read_csv('Data/progresses.csv')# Читаем файл progresses.csv
df_progresses.rename(columns={'id':'progress_id'}, inplace=True)

df_progress_phases = pd.read_csv('Data/progress_phases.csv', parse_dates = ['start_date','finish_date'])# Читаем файл progress_phases.csv

# Студенты
# Столбец id_ здесь лишний, а столбец id - переименовываем в student_id
df_students = pd.read_csv('Data/students.csv', usecols=['id', 'city', 'birthday'])
df_students.rename(columns={'id':'student_id'}, inplace=True)

```

```python
df_main = df_progresses.merge(df_progress_phases, on='progress_id', how='left')
df_main = df_main.merge(df_courses, on='course_id', how='right')
df_main = df_main.merge(df_course_contents, on=['course_id', 'module_number', 'lesson_number'], how='outer')
df_main.head(2)
```

```python
# Общее количество курсов в датасете
print('Общее количество курсов в датасете:', df_main.course_id.nunique(), '\n')
```

```python
# Количество модулей на каждом курсе
print('Количество модулей на каждом курсе:')
df_main.groupby('title')['module_number'].nunique().sort_values(ascending=False)
```

```python
# Количество уроков на каждом модуле каждого курса.
les_by_mod_by_courses = df_main.groupby(['title','module_number'])['lesson_token'].nunique().unstack(level=0)
les_by_mod_by_courses
```

```python
# Медианное количество уроков в каждом модуле каждого курса.
print('Медианное количество уроков в модуле для каждого курса:')
les_by_mod_by_courses.apply('median', axis=0)
```

```python
#количество учеников на каждом курсе
df_main.student_id.nunique()
```

```python
# Количество учеников на каждом курсе
students_count_by_course = df_main.groupby('title').student_id.nunique().sort_values(ascending=False)
students_count_by_course
```

```python
df_main_ages = df_main.merge(df_students.dropna(subset=['birthday']), on='student_id', how='right')
```

```python
temp = pd.to_datetime(df_students.birthday.dropna(), errors='coerce')
print('Все некорректные даты рождения:\n',df_students.birthday.dropna()[temp.isna()], sep='')
```

```python
df_main_ages['birthday'] = pd.to_datetime(df_main_ages['birthday'], errors='coerce')
#df_main_ages['birthday']=df_main_ages['birthday'].to_pydatetime()#-df_main_ages['start_date'].dt.date

df_main_ages['start_date'] = pd.to_datetime(df_main_ages['start_date'], errors='coerce')
df_main_ages['finish_date'] = pd.to_datetime(df_main_ages['finish_date'], errors='coerce')
#df_main_ages['start_date'].dt.to_pydatetime()

df_main_ages['start_date'] = df_main_ages['start_date'].dt.date

now = pd.to_datetime('now')
df_main_ages['age'] = (now - df_main_ages['birthday']).astype('<m8[Y]') 

# отфильтруем очевидно большой и отрицательный возраст

df_main_ages = df_main_ages.loc[(df_main_ages['age'] <= 96) & (df_main_ages['age'] >= 12)]

# Данные о возрасте студентов
print('Данные о возрасте студентов:\n',\
      df_main_ages.age.agg({'Максимальный возраст': 'max', 'Минимальный возраст': 'min',\
                            'Средний возраст':'mean', 'Медианный возраст': 'median'}), sep='')
```

```python
# Данные о возрасте студентов по каждому из курсов
# Группируем по курсам и сразу считаем все 4 величины
ages_by_course = df_main_ages.groupby('title').agg({'age':['max','min','mean','median']})
ages_by_course.head(30)
# Убираем иерархический индекс у столбцов.
ages_by_course.columns = ages_by_course.columns.droplevel(0)

# Добавляем префикс age к именам столбцов, для лучшей читаемости.
ages_by_course.rename(columns={'max':'age_max', 'min':'age_min', 'mean':'age_mean', 'median':'age_median'}, inplace=True)
print('Данные о возрасте студентов по каждому курсу:')
ages_by_course.sort_values(by='age_median', ascending=False)
```

```python
# Столбчатая диаграмма отражающая количество студентов на каждом из курсов:

plt.figure(figsize=(20,10))
ax = plt.bar(students_count_by_course.index, students_count_by_course)
# Подписи столбцов поворачиваем, выравниваем по правому краю и привязываем к столбцу
plt.title('Количество студентов по курсам', fontsize=25)
plt.xticks(rotation=30, ha='right', rotation_mode='anchor', fontsize=14)
plt.yticks(fontsize=15)
plt.xlabel(xlabel='Название курса', fontsize = 20)
plt.ylabel(ylabel='Количество студентов', fontsize=20)
plt.show()
```

```python
#df_fch separate dataframe for chart(fch) 

df_fch = df_main.groupby('title').agg({'student_id':'nunique', 'field':'first'}).\
                sort_values(by='student_id', ascending=False)
# Let's make dict for colors for each area.

mapper = sns.color_palette('bright',df_fch.field.unique().size)
color_dict = {}
for num, field in enumerate(df_fch.field.unique()):
    color_dict[field] = mapper[num]

my_palette = []
for element in df_fch.field:
    my_palette.append(color_dict[element])
my_palette

plt.figure(figsize=(20,10))
# Альфа 0.9 
students_hbar = sns.barplot(x='student_id', y=df_fch.index, data=df_fch, palette=my_palette, alpha=0.9)

# Отрисуем вертикальную линию по медианному количеству студентов.
fl = plt.axvline(df_fch.student_id.median(), ls='--', lw=3, color='purple', alpha=0.9, label='Медианное количество студентов')
# Добавим для неё отдельный xtick
xt = students_hbar.get_xticks() 
xt = np.append(xt,df_fch.student_id.median())
students_hbar.set_xticks(xt)

# И отдельную легенду для этой линии.
first_legend = plt.legend(handles=[fl], loc=7, fontsize = 20, shadow=True)
# Теперь добавим эту легенду в текущий axes
plt.gca().add_artist(first_legend)

# Нарисуем вручную прямоугольники для легенды и присвоим им цвета из нашего словаря цветов.
handles = [plt.Rectangle((0,0),1,1, color=color_dict[key], alpha=.9) for key in color_dict.keys()]
# Отрисовываем вторую легенду.
plt.legend(handles, color_dict.keys(), fontsize=20, title='Field', title_fontsize=25, shadow=True)

# Настроим подписи графика, осей, меток.
plt.title('Количество студентов по курсам', fontsize=25)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.xlabel('Количество студентов', fontsize=20)
plt.ylabel(None)

# Убираем рамку и отрисовываем вертикальные линии сетки, для лучшей читаемости диаграммы.
plt.box(False)
plt.grid(axis='x', alpha=0.3, ls='-', lw=2)
plt.show()
```

```python
#3.1Рассчитать минимальное, максимальное, среднее, медианное время прохождения каждого модуля 
#(разность между временем начала и окончания выполнения домашней работы) для каждого курса. 
#Если домашних заданий в модуле несколько, то считать разность между временем начала выполнения первой домашней работы и 
#временем окончания выполнения последней домашней работы в модуле

condition = (df_main['start_date'] < df_main['finish_date']) & (df_main['is_homework']==True)
# Отбросим записи с отсутствующими значениями в столбцах с датами начала и завершения домашней работы.
df3_1 = df_main[condition].dropna(subset=['start_date', 'finish_date'])
# Для каждого студента определим самую раннуюю дату начала домашней работы и самую позднюю дату завершения
df3_1 = df3_1.groupby(['title','module_number', 'module_title','student_id'])\
                        .agg({'start_date':'min', 'finish_date':'max'}).reset_index()
# Создадим столбец с временем выполнения домашней работы
df3_1['homework_time'] = df3_1['finish_date'] - df3_1['start_date']
# Сгруппируем датафрейм по курсам, названию и номеру модуля и сразу посчитаем сразу max,min,mean и median
# Тут есть небольшая особенность. Дело в том что функция mean() и median() в python 
# может работать и для объектов типа timedelta. Но в pandas groupby.agg() реализован на cython так, что 
# mean и median работают только с numeric типом данных. Я в данном случае преобразовал timedelta к int64,
# потом посчитал все нужные параметры и преобразовал обратно к timedelta.
df3_1['homework_time'] = df3_1['homework_time'].astype('int64')
df3_1 = df3_1.groupby(['title','module_number','module_title'])\
                        .agg({'homework_time':['min','max','mean','median']}).astype('timedelta64[ns]')
# Сбрасываем мультииндекс у столбцов и уровень module_title у строк
df3_1 = df3_1.droplevel(0,axis='columns').reset_index(level=2)
df3_1.head(15)
```

```python
#3.2
fig, ax = plt.subplots(5,3, figsize=(20,25))
# Создаём список курсов. Можно и из исходного датафрейма их вытащить.
titles = df3_1.index.get_level_values(0).unique()
counter = 0

for i in range(5):
    for j in range(3):
# Создаём временный датафрейм для конкретного курса.         
        current_df = df3_1.loc[titles[counter]]['median']
# Реиндексируем временный датафрейм. Дело в том, что по умолчанию plot соединит все точки линиями,
# даже те, где отсутствуют данные. Это может привести к неверной интерпретации графиков.
# Может сложиться впечатление, что медианное время выполнения задания плавно нарастает или убывает,
# хотя это во многих случаях совсем не так.
# Если же значения будут NaN - то plot такие части просто не соединит. Это то, что нам нужно.
        current_df_ri = current_df.reindex(range(current_df.index.astype('int').max()+1))
# Отрисовываем графики для текущего курса. Маркеры выставляем, для того, чтобы было видно те самые 
# "отдельно стоящие" модули. Пунктиром обозначен график без реиндексирования.
# По умолчанию matplotlib timedelta интерпретирует в миллисекундах, поэтому вручную перевожу период в секунды
# и делю на 86400(количество секунд в сутках)
        ax[i,j].plot(current_df.index, current_df.dt.total_seconds()/86400,\
                     ls='--', alpha=0.5, lw=2, marker='o', markersize=7, label='not_reindexed', color='darkorange')
        ax[i,j].plot(current_df_ri.index, current_df_ri.dt.total_seconds()/86400,\
                     lw=2, marker='o', markersize=7, label='reindexed', color='tab:blue')
# Настраиваем оси, подписи, сетку(для лучшей читаемости).
        ax[i,j].set_xticks(current_df_ri.index)
        ax[i,j].tick_params(labelsize=14)
        ax[i,j].set_xlabel('Номер модуля', fontsize=14)
        ax[i,0].set_ylabel('Медианное время\n выполнения работы (дней)', fontsize=14)
        ax[i,j].grid(alpha=0.2)
        ax[i,j].set_title(titles[counter], fontsize=18)
        ax[i,j].legend(fontsize=12)
        counter += 1
# tight_layout увеличит расстояние между axes, таким образом подписи не будут наезжать друг на друга.
# Всё то же самое можно сделать вручную, назначив вертикальное и горизонтальное расстояние между графиками.
plt.tight_layout()
plt.show()
```

```python
#3.3 (p) Чтобы выявить сезонность, посчитать медианное время выполнения домашней работы по месяцам (12 месяцев, январь-декабрь) 
# для каждого курса.
# Сначала почистим датафрейм. Оставим только домашние задания и отбросим те записи,
# где дата начала домашнего задания больше чем дата его окончания. На такие записи потом посмотрим отдельно.
# Чтобы код не загромождать запишем условие фильтрации отдельно.
condition = (df_main['start_date'] < df_main['finish_date']) & (df_main['is_homework']==True)
# Отбросим записи с отсутствующими значениями в столбцах с датами начала и завершения домашней работы.
df3_3 = df_main[condition].dropna(subset=['start_date', 'finish_date'])
# Создадим два столбца, куда запишем только месяцы старта и завершения работы. Для последующей сотрировки
df3_3[['start_month','finish_month']] = df3_3[['start_date','finish_date']].applymap(lambda x: x.month)
# Посчитаем время выполнения домашней работы и запишем в отдельный столбец.
df3_3['homework_time_hours'] = df3_3['finish_date'] - df3_3['start_date']
# Сделаем два варианта, сгруппировав по месяцам начала выполнения домашней работы и по месяцам завершения.
# Сравним как они отличаются на графиках.
df3_3_start = df3_3.groupby(['title','start_month'])['homework_time_hours'].median(numeric_only=None)
df3_3_finish = df3_3.groupby(['title','finish_month'])['homework_time_hours'].median(numeric_only=None)
```

```python
fig, ax = plt.subplots(5,3, figsize=(20,25))
# Создаём список курсов, для итерирования по нему
titles = df3_3_start.index.get_level_values(0).unique()
counter = 0
for i in range(5):
    for j in range(3):
# Отрисовываем графики по данным отсортированным по месяцу начала выполнения работы и по месяцу завершения.
# Реиндексацию здесь не делал, потому что эти графики построил только для проверки различий.
# По умолчанию matplotlib timedelta интерпретирует в миллисекундах, поэтому вручную перевожу период в секунды
# и делю на 86400(количество секунд в сутках)
        ax[i,j].plot(df3_3_start.loc[titles[counter]].index,\
                     df3_3_start.loc[titles[counter]].dt.total_seconds()/86400, label='By start', lw=2)
        ax[i,j].plot(df3_3_finish.loc[titles[counter]].index,\
                     df3_3_finish.loc[titles[counter]].dt.total_seconds()/86400, label='By finish', lw=2)
# Настраиваем параметры осей, подписи, сетку.
        ax[i,j].set_xticks(df3_3_start.loc[titles[counter]].index)
        ax[i,j].tick_params(labelsize=12)
        ax[i,j].grid(alpha=0.2)
        ax[i,0].set_ylabel('Медианное время\n выполнения работы (дней)', fontsize=14)
        ax[i,j].set_xlabel('Месяцы', fontsize=14)
        ax[i,j].set_title(titles[counter], fontsize=18)
        ax[i,j].legend()
        counter += 1
# tight_layout увеличит расстояние между axes, таким образом подписи не будут наезжать друг на друга.
plt.tight_layout()
plt.show()
```

```python
# Создадим датафрейм с усредненными значениями и развернём его с помощью unstack
# Таким образом сразу получаем единую шкалу для всех курсов.
df3_3_mean = (df3_3_start+df3_3_finish)/2
df3_3_mean = df3_3_mean.unstack(level= 0)

# Создадим словарь для цветов
# В matplotlib по-умолчанию используется циклическая палитра из 8 цветов, 
# поэтому при 15 графиках цвета будут дублироваться. 
# Чтобы цвета всех графиков были разные я и использую созданный словарь.
colors = sns.color_palette('tab20', 15)
courses = df3_3_mean.columns
color_dict = {}
for color,course in enumerate(courses):
    color_dict[course] = colors[color]

plt.figure(figsize=(20,20))
for title in df3_3_mean.columns:
# По умолчанию matplotlib timedelta интерпретирует в миллисекундах, поэтому вручную перевожу период в секунды
# и делю на 86400(количество секунд в сутках)
    plt.plot(df3_3_mean.index, df3_3_mean[title].dt.total_seconds()/86400, label=title, lw=2, color=color_dict[title])
# Настраиваем оси, подписи, сетку, легенду
plt.xticks(df3_3_mean.index, fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Месяцы', fontsize=18)
plt.ylabel('Время выполнения (дней)', fontsize=18)
plt.title('Медианное время выполнения домашней работы по месяцам', fontsize=25)
plt.grid(axis='both', alpha=0.2, ls='--', lw=2)
leg = plt.legend(fontsize=14)
# Линии слишком тонкие в базе, поэтому увеличиваю их толщину в цикле, так лучше читается.
[line.set_linewidth(4) for line in leg.get_lines()]
plt.box(False)
plt.show()
```

## 4. Расчет конверсии

```python
# 4.1
# По аналогии со вторым блоком курсового проекта нужно сначала получить список всех первых уроков 
# c домашними заданиями в каждоммодуле и список всех последних уроков с домашними заданиями в каждом модуле.
# Нужны два списка, их потом будем использовать при объединении как фильтр.

# Собираем датафрейм с номерами в которых впервые появляется домашнее задание в модуле. 
# Сначала отфильтровываем только те строки, которые относятся к домашним работам.
# затем группируем по курсам и модулям и выбираем уроки с минимальным номером.
first_lessons = df_main.loc[df_main['is_homework'] == True].groupby(['title','module_number'])\
                            .agg({'lesson_number':'min'}).reset_index()

# Теперь соберём такой же датафрейм для последних уроков с домашним заданием по модулям
last_lessons = df_main.loc[df_main['is_homework'] == True].groupby(['title','module_number'])\
                            .agg({'lesson_number':'max'}).reset_index()

# Для подсчёта завершивших модуль нам нужны только записи со статусом done.
students_done = df_main.loc[df_main['status'] == 'done'].dropna(subset=['start_date','finish_date'])\
                            .groupby(['title','module_number','module_title','lesson_number'])\
                            .agg(studs_done =('student_id','nunique')).reset_index()

# # А для подсчёта начавших модуль нам нужны будут записи со всеми статусами.
students_start = df_main.dropna(subset=['start_date']).groupby(['title','module_number','module_title','lesson_number'])\
                                .agg(studs_start =('student_id','nunique')).reset_index()

# С помощью merge фильтруем нужные нам данные. Получаем два датафрейма с количеством студентов.
df4_1_done = last_lessons.merge(students_done, on=['title','module_number','lesson_number'], how='left').drop(columns=['lesson_number'])
df4_1_start = first_lessons.merge(students_start, on=['title','module_number','lesson_number'], how='left' ).drop(columns=['lesson_number'])

# Объединим их в общий датафрейм
df4_1 = df4_1_start.merge(df4_1_done, on=['title','module_number','module_title'], how='left')
# Теперь сдвинем на одну строку вниз весь столбец с количеством студентов закончивших модуль.
df4_1['studs_done'] = df4_1['studs_done'].shift(1)
# # Первый модуль по каждому курсу нам не нужен, конверсия может быть только между модулями.
# # Соответственно число конверсий будет всегда на 1 меньше, чем число модулей.
# # В цикле убираем первые модули в каждому курсе
for name in df4_1.title.unique():
    df4_1.drop(index=(df4_1.loc[df4_1['title'] == name].module_number.idxmin()), inplace=True)

# Наконец считаем конверсию. Записываем в отдельный столбец
df4_1['conversion'] = df4_1['studs_start']/df4_1['studs_done']
```

```python
# Один общий barplot
# Создем список подписей к оси Х. Поскольку конверсия в данном случае определяет именно переход из одного модуля в другой,
# число конверсий всегда будет меньше числа модулей на единицу. Именно поэтому я удалил первые строки по каждому курсу из
# датафрейма. Чтобы было понятно, что собой представляет ось х, я создаю подписи отражающие из какого модуля в какой
# показана конверсия.
labels_list=[]
for i in range(1,21):
    labels_list.append(f'{i}-->{i+1}')
    
plt.figure(figsize=(20,10))
# Строим сводную столбчатую диаграмму на базе датафрейма. Использую seaborn потому, что здесь в barplot можно 
# категоризировать bar-ы используя параметр hue. Так гораздо проще, чем писать дополнительный кусок кода, 
# для сдвига столбцов в обычном plt.bar
# Добавил окантовку толщиной 0.4 к каждому бару, без неё они сливаются в единые фигуры, сложно ориентироваться.
consolidated_barplot = sns.barplot(x='module_number', y='conversion', hue='title', data=df4_1,\
                                  palette=sns.color_palette('tab20', 15), saturation=0.7, linewidth=.4,  edgecolor=".2")
# Настраиваем оси, подписи и т.д.
plt.xlabel('Между какими модулями конверсия', fontsize=16)
plt.ylabel('Конверсия', fontsize=16)
# Для подписей оси х используем созданный список.
ax = plt.gca()
ax.set_xticklabels(labels_list, rotation=30)
ax.tick_params(labelsize=14)
# Отрисовываем легенду
plt.legend(loc='upper right', fontsize=12)
plt.show()
```
